```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 📊 Sales by Coffee Type (e.g., top-selling drinks)
# 1---> by using bar graph
df["coffee_name"].value_counts()


```

```python
df["coffee_name"].value_counts().plot(kind="bar")
# plt.savefig("coffee_bar.png")
```

```python
# 2.using pie
df["coffee_name"].value_counts().plot(kind="pie",autopct="%.2f")
# plt.savefig("coffee_pie.png")
```

```python
#which coffee have more revenue?
df.groupby("coffee_name")["money"].sum().plot(kind="bar",color=["red","green","blue"])
plt.title("coffee sales revenue by coffee name")
plt.xlabel("coffee name")
plt.ylabel("revenue")
```

```python
df.groupby("coffee_name")["hour_of_day"].sum()
```

```python
df.groupby("hour_of_day")["money"].sum().plot(kind="line")
plt.xlabel("hour_of_day")
plt.ylabel("money")
plt.savefig("hour_bar.png")
```
